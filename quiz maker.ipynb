```python
import tkinter as tk
from tkinter import messagebox

# Questions, options, and correct answers
questions = [
    "1. What is the capital of India?",
    "2. Which one is a programming language?",
    "3. What is the largest planet?",
    "4. Which animal is known as the King of the Jungle?",
    "5. What is H2O?",
]

options = [
    ["Delhi", "Mumbai", "Kolkata"],
    ["Python", "Banana", "Chair"],
    ["Earth", "Jupiter", "Mars"],
    ["Elephant", "Lion", "Tiger"],
    ["Water", "Oxygen", "Hydrogen"]
]

answers = [0, 0, 1, 1, 0]  # correct option index

current_q = 0
score = 0
answer_shown = False

def next_question():
    global current_q, score, answer_shown

    selected = selected_option.get()

    # If answer not shown yet, show it
    if not answer_shown:
        if selected == -1:
            messagebox.showwarning("Warning", "Please select an option!")
            return
        
        correct_ans_text = options[current_q][answers[current_q]]

        if selected == answers[current_q]:
            result_label.config(text=f"Correct! 🎉\nAnswer: {correct_ans_text}", fg="green")
            score += 1
        else:
            result_label.config(text=f"Wrong ❌\nCorrect Answer: {correct_ans_text}", fg="red")

        answer_shown = True
        next_btn.config(text="Next")
        return

    answer_shown = False
    result_label.config(text="")
    selected_option.set(-1)

    current_q += 1

    if current_q == len(questions):
        messagebox.showinfo("Quiz Completed", f"Your Score: {score}/5")
        root.destroy()
        return

    load_question()
    next_btn.config(text="Submit")

def load_question():
    question_label.config(text=questions[current_q])

    for i in range(3):
        option_buttons[i].config(text=options[current_q][i])


root = tk.Tk()
root.title("Quiz Maker With Answers")
root.geometry("420x350")

question_label = tk.Label(root, text="", font=("Arial", 14), wraplength=380)
question_label.pack(pady=20)

selected_option = tk.IntVar()
selected_option.set(-1)

option_buttons = []
for i in range(3):
    rb = tk.Radiobutton(root, text="", variable=selected_option, value=i, font=("Arial", 12))
    rb.pack(anchor="w", padx=40)
    option_buttons.append(rb)

result_label = tk.Label(root, text="", font=("Arial", 13))
result_label.pack(pady=10)

next_btn = tk.Button(root, text="Submit", command=next_question, font=("Arial", 12))
next_btn.pack(pady=20)

load_question()

root.mainloop()

```
